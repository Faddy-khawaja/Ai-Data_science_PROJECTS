Q14. Amazon — Write a function that returns the factorial of a number using a while loop.


```
def fact_user(n):
  i = 1
  fact = 1
  while i <= n:
    fact = fact * i
    i = i+1
  return fact

print(fact_user(5))
print(fact_user(0))
```

**Q15. Microsoft** — Write a function that takes a string and counts **how many uppercase and lowercase letters** it has.

```
def count_upper_lower(string):
    upper = 0
    lower = 0

    for char in string:
        if char.isupper():
            upper += 1
        if char.islower():
            lower += 1

    print(f"Uppercase: {upper}")
    print(f"Lowercase: {lower}")

count_upper_lower("Hello World!")
```

**Q16. Meta** — Write a function that accepts a list of numbers and returns a **new list with only unique elements** (no duplicates).

```
def new_list(number):
    unique_list = []
    for num in number:
        if num not in unique_list:
            unique_list.append(num)
    return unique_list

# This code should be outside the function
number = [1, 1, 2, 3, 4, 4, 5]
unique_list = new_list(number)
print(unique_list)
```

**Q17. Google** — Write a function to check if a number is **prime or not**.






```
def is_prime(n):


    if n < 2:
        return False


    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

print(is_prime(7))
print(is_prime(10))
print(is_prime(2))
print(is_prime(1))
```

**Q20. Amazon** — Write a Python function to find the **maximum product of two numbers in a list** (e.g., `[2, 3, 5, 7] → 35`).

```
def max_product(numbers):
    numbers.sort()
    return max(numbers[-1] * numbers[-2], numbers[0] * numbers[1])


print(max_product([2, 3, 5, 7]))
print(max_product([-10, -3, 5, 2, 7]))
print(max_product([-5, -4, 1, 2, 3]))
print(max_product([1, 2]))
```
