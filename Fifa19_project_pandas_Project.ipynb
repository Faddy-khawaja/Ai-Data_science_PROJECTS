```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("fifa19.csv")
df.head()
```

Find top 10 highest-paid **players**

```
top_10 = df.sort_values(by='Wage', ascending=False).head(10)
top_10[['Name', 'Wage']]
```

- Most valuable players by position  
**bold text**

```
top3_by_pos = df.sort_values('Value', ascending=False).groupby('Position').head(3)
top3_by_pos[['Position', 'Name', 'Value']]
```

- Plot distribution of 'Potential'  


```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['Potential'], bins=20, kde=True)
plt.title("Distribution of Player Potential")
plt.xlabel("Potential")
plt.ylabel("Frequency")
plt.show()
```

- Create Age Group feature using cut()  
**bold text**

```
bins = [0, 18, 25, 30, 35, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Senior', 'Veteran']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
df[['Age', 'Age_Group']].head()
```
