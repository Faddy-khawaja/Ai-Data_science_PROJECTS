Analyze Hypertension, Diabetes & Alcoholism effect	Use **groupby**

```
df.groupby("Hipertension")["No-show"].value_counts(normalize=True)
```

```
df.groupby('Diabetes')['No-show'].value_counts(normalize=True)
```

```
df.groupby('Alcoholism')['No-show'].value_counts(normalize=True)
```

```
df['No-show'] = df['No-show'].map({'Yes': 1, 'No': 0})
```

Neighborhood vs No-Show % graph	Use **barplot**

```
import seaborn as sns
import matplotlib.pyplot as plt

neigh_show = df.groupby('Neighbourhood')['No-show'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=neigh_show.index, y=neigh_show.values)
plt.xticks(rotation=90)
plt.ylabel("No-Show Percentage")
plt.title("No-Show % by Neighborhood")
plt.show()
```

Final Summary paragraph	Key reasons behind no-**show**

Our data shows that some health problems like high blood pressure, diabetes, and drinking too much alcohol can make patients miss their doctor appointments more often. This might be because they feel sicker or are less motivated to go.

We also saw that some neighborhoods have more missed appointments than others. Things like how much money people have or how easy it is to get to the clinic can play a part.

Another big reason is waiting time. If patients schedule an appointment many days before the actual date, they are more likely to not show up.

So, in simple terms, the main reasons people miss appointments are:

Long waits between booking and the appointment.

Where they live.

Their health problems.**bold text**
