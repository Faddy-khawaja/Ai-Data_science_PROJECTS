Project Title:
Patient Health Monitoring & Appointment Management System (Core Python Project)
Project Description:
A small clinic wants to simplify its daily workflow by digitizing patient records, tracking vitals, and
managing appointments. Currently, the clinic uses manual paper-based forms, which leads to
lost records, double appointments, inaccurate vitals tracking, and difficulty checking patient
history.
Your task is to build a Python console-based application (no external libraries) to help clinic
staff manage patient data efficiently.**bold text**

```
# ===== Patient Health Monitoring System =====

patients = {}
vitals = {}
appointments = []

# ---------- File Load ----------
def load_data():
    global patients, vitals, appointments
    try:
        patients = eval(open("patients.txt").read())
    except:
        patients = {}
    try:
        vitals = eval(open("vitals.txt").read())
    except:
        vitals = {}
    try:
        appointments = eval(open("appointments.txt").read())
    except:
        appointments = []

# ---------- File Save ----------
def save_data():
    open("patients.txt", "w").write(str(patients))
    open("vitals.txt", "w").write(str(vitals))
    open("appointments.txt", "w").write(str(appointments))

# ---------- Patient ----------
def add_patient():
    try:
        name = input("Name: ")
        age = int(input("Age: "))
        gender = input("Gender: ")
        phone = input("Contact: ")
        cond = input("Conditions (comma): ").split(",")
        patients[name] = {"age": age, "gender": gender, "phone": phone, "cond": cond}
        print("Patient Added")
    except:
        print("Invalid Input")

def view_patients():
    for k, v in patients.items():
        print(k, v)

# ---------- Vitals ----------
def add_vitals():
    name = input("Patient Name: ")
    if name not in patients:
        print("Patient not found")
        return
    try:
        temp = float(input("Temp: "))
        bp = input("BP: ")
        hr = int(input("Heart Rate: "))
        wt = float(input("Weight: "))
        date = input("Date: ")
        vitals.setdefault(name, []).append(
            {"date": date, "temp": temp, "bp": bp, "hr": hr, "wt": wt}
        )
        print("Vitals Saved")
    except:
        print("Invalid vitals")

def view_vitals():
    name = input("Patient Name: ")
    print(vitals.get(name, "No vitals"))

# ---------- Appointment ----------
def book_appointment():
    name = input("Patient Name: ")
    date = input("Date: ")
    time = input("Time: ")
    doctor = input("Doctor: ")
    for a in appointments:
        if a["date"] == date and a["time"] == time:
            print("Slot already booked")
            return
    appointments.append(
        {"name": name, "date": date, "time": time, "doctor": doctor}
    )
    print("Appointment Booked")

def view_appointments():
    for a in appointments:
        print(a)

def cancel_appointment():
    name = input("Patient Name: ")
    for a in appointments:
        if a["name"] == name:
            appointments.remove(a)
            print("Cancelled")
            return
    print("Not found")

# ---------- Main ----------
load_data()
while True:
    print("\n1.Add Patient 2.View Patients 3.Add Vitals 4.View Vitals")
    print("5.Book Appointment 6.View Appointments 7.Cancel 8.Save & Exit")
    ch = input("Choice: ")

    if ch == "1": add_patient()
    elif ch == "2": view_patients()
    elif ch == "3": add_vitals()
    elif ch == "4": view_vitals()
    elif ch == "5": book_appointment()
    elif ch == "6": view_appointments()
    elif ch == "7": cancel_appointment()
    elif ch == "8":
        save_data()
        print("Saved & Exit")
        break
    else:
        print("Wrong choice")
```
